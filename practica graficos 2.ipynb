desafio1: solucionar lo de los acumulados para el stacking usando el metodo numpy.cumsum

desafio2: buscar una tabla con informacin y agregar los valores a un grafico de barras apiladas usando un for


```python
import numpy as np
import matplotlib.pyplot as plt

categorias=['A', 'B', 'C', 'D', 'E']
valores1=np.array([25, 30, 15, 45, 30])
valores2=np.array([30, 45, 25, 55, 40])
valores3=np.array([15, 10, 5, 30, 20])
error= [3, 4, 2, 5, 3]
colors=['blue', 'red']
#si cada serie tuviera errores distintos, como suele ser, se hace un array de errores y se trabaja igual

data=np.array([valores1,valores2,valores3])
#print(data)
bottom=np.cumsum(data, axis=0) #la funcion cumsum crea un array resultado de ir sumando los valores de cada posicion
#print(bottom)

index=np.arange(5)
width=0.5

plt.bar(index, data[0], width=width, yerr=error[0], label= 'valores 1', color='green') #el primero separado permite que se apile sobre 0
for j in range(1, data.shape[0]): #recorre los datos desde 1 hasta el final del tamao de los datos, en este caso 5
    plt.bar(index,data[j],width=width, yerr=error[j-1], label='valores {}'.format(j+1), color=colors[j-1], bottom=bottom[j-1])
#se van recorriendo los datos y agregando los valores al grafico, sobre el acumulado del punto anterior
    
plt.ylabel('valores')
plt.xlabel('categorias')
plt.xticks(index, categorias) #para centrar las categorias respecto al indice que es donde comienza la primer barra
plt.legend(loc='upper left')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)

plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap='jet')
C = plt.contour(X, Y, f(X, Y), 8, colors='black', linewidth=.5)
```
